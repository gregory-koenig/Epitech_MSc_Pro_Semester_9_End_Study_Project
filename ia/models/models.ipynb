```python
import random
import json
```

```python
ast = open("ast.json", "r")
json_ast = json.load(ast)
```

```python
ast_str = ""
for line in json_ast["ast_vec"]:
    ast_str += f"{line}, "

ast_str = ast_str [:-2]

print(ast_str)
```

```python
ast_vec = []
tree_depth = []
nb_var = []
nb_loops = []
nb_nested_loops = []
nb_leaf = []
algo_vec = []
algo = [0,0,0,0,1,2,3]

for _ in range(10000):
    depth = random.randint(1,200)
    loops = random.randint(0, depth)
    var = random.randint(0, depth)
    nested_loops = random.randint(0, loops)
    leaf = random.randint(depth, depth*4)
    algo_sample = random.randint(1, 210)
    
    ast_vec.append(ast_str)
    tree_depth.append(depth)
    nb_var.append(var)
    nb_loops.append(loops)
    nb_nested_loops.append(nested_loops)
    nb_leaf.append(leaf)
    algo_vec.append(algo_sample)
```

```python
train["ast_vec"] = ast_vec
train["tree_depth"] = tree_depth
train["nb_var"] = nb_var
train["nb_loops"] = nb_loops
train["nb_nested_loops"] = nb_nested_loops
train["nb_leaf"] = nb_leaf
train["algo_vec"] = algo_vec
```
